# Sayılar Teorisi

Bölünebilirlik, Öklid algoritması, modüler aritmetik ve asal sayılar.

## 1. En Büyük Ortak Bölen

**Öklid algoritması** $\\gcd(a, b)$ değerini verimli biçimde hesaplar.

```python
def ebob(a, b):
    while b:
        a, b = b, a % b
    return a

def ekok(a, b):
    return a * b // ebob(a, b)

print("ebob(48, 36) =", ebob(48, 36))
print("ekok(4, 6)   =", ekok(4, 6))
```

## 2. Genişletilmiş Öklid Algoritması

$ax + by = \\gcd(a, b)$ olacak şekilde $x, y$ tam sayılarını bulur (Bézout özdeşliği).

```python
def genisletilmis_ebob(a, b):
    if b == 0:
        return a, 1, 0
    g, x1, y1 = genisletilmis_ebob(b, a % b)
    return g, y1, x1 - (a // b) * y1

g, x, y = genisletilmis_ebob(48, 36)
print(f"ebob = {g}, x = {x}, y = {y}")
print(f"Kontrol: 48·{x} + 36·{y} = {48*x + 36*y}")
```

## 3. Modüler Aritmetik

Modüler üs alma ve modüler tersler kriptografide merkezi öneme sahiptir.

```python
def mod_ters(a, m):
    g, x, _ = genisletilmis_ebob(a, m)
    if g != 1:
        return None  # ters yoktur
    return x % m

print("3^4 mod 5 =", pow(3, 4, 5))
print("3'ün mod 7'ye göre tersi =", mod_ters(3, 7))
```

## 4. Asallık ve Eratosthenes Kalburu

$n$'ye kadar olan tüm asal sayıları verimli biçimde üretir.

```python
def kalbur(n):
    asal_mi = [True] * (n + 1)
    asal_mi[0] = asal_mi[1] = False
    for i in range(2, int(n ** 0.5) + 1):
        if asal_mi[i]:
            for j in range(i * i, n + 1, i):
                asal_mi[j] = False
    return [i for i, p in enumerate(asal_mi) if p]

print("50'ye kadar asallar:", kalbur(50))
```
