# Bağıntılar ve Fonksiyonlar

İkili bağıntılar, özellikler (yansımalı, simetrik, geçişli), denklik bağıntıları ve fonksiyonlar.

## 1. İkili Bağıntılar

Bir $A$ kümesi üzerindeki ikili bağıntı $R$, $A \\times A$'nın bir alt kümesidir. Bunu ikililer kümesi olarak gösteririz.

```python
A = {1, 2, 3}
R = {(1, 1), (2, 2), (3, 3), (1, 2), (2, 1)}

def yansimali_mi(A, R):
    return all((a, a) in R for a in A)

def simetrik_mi(R):
    return all((b, a) in R for (a, b) in R)

def gecisli_mi(R):
    return all((a, c) in R for (a, b) in R for (c, d) in R if b == c)

print("Yansımalı: ", yansimali_mi(A, R))
print("Simetrik:  ", simetrik_mi(R))
print("Geçişli:   ", gecisli_mi(R))
```

## 2. Denklik Bağıntıları

Bir **denklik bağıntısı** yansımalı, simetrik ve geçişlidir. Kümeyi denklik sınıflarına böler.

```python
def denklik_mi(A, R):
    return yansimali_mi(A, R) and simetrik_mi(R) and gecisli_mi(R)

def denklik_siniflari(A, R):
    siniflar = []
    gorulen = set()
    for a in A:
        if a in gorulen:
            continue
        sinif = {b for b in A if (a, b) in R}
        siniflar.append(sinif)
        gorulen |= sinif
    return siniflar

print("Denklik bağıntısı mı:", denklik_mi(A, R))
print("Denklik sınıfları:", denklik_siniflari(A, R))
```

## 3. Fonksiyonlar

Bir **fonksiyon** $f: A \\to B$, $A$'nın her elemanına $B$'nin tam olarak bir elemanını atar. Birebir (injektif), örten (surjektif) ve birebir-örten (bijektif) fonksiyonları inceliyoruz.

```python
def birebir_mi(f, tanim_kumesi):
    goruntuler = [f(x) for x in tanim_kumesi]
    return len(goruntuler) == len(set(goruntuler))

def orten_mi(f, tanim_kumesi, deger_kumesi):
    return set(f(x) for x in tanim_kumesi) == set(deger_kumesi)

tanim_kumesi = [1, 2, 3]
deger_kumesi = [2, 4, 6]
f = lambda x: 2 * x

print("Birebir:        ", birebir_mi(f, tanim_kumesi))
print("Örten:          ", orten_mi(f, tanim_kumesi, deger_kumesi))
print("Birebir-örten:  ", birebir_mi(f, tanim_kumesi) and orten_mi(f, tanim_kumesi, deger_kumesi))
```
