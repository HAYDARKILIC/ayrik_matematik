# Kombinatorik

Sayma ilkeleri, permütasyonlar, kombinasyonlar, binom teoremi ve güvercin yuvası ilkesi.

## 1. Permütasyonlar ve Kombinasyonlar

$P(n, r) = \\frac{n!}{(n-r)!}$ sıralı dizilişleri sayar; $C(n, r) = \\frac{n!}{r!(n-r)!}$ sırasız seçimleri sayar.

```python
from math import factorial, comb, perm

def permutasyon_sayisi(n, r):
    return factorial(n) // factorial(n - r)

def kombinasyon_sayisi(n, r):
    return factorial(n) // (factorial(r) * factorial(n - r))

print("P(5, 2) =", permutasyon_sayisi(5, 2), "| yerleşik:", perm(5, 2))
print("C(5, 2) =", kombinasyon_sayisi(5, 2), "| yerleşik:", comb(5, 2))
```

## 2. Binom Teoremi

$(x + y)^n = \\sum_{k=0}^{n} \\binom{n}{k} x^{n-k} y^k$

```python
def binom_acilimi(n):
    terimler = []
    for k in range(n + 1):
        katsayi = comb(n, k)
        terimler.append(f"{katsayi}·x^{n-k}·y^{k}")
    return " + ".join(terimler)

print(f"(x + y)^4 = {binom_acilimi(4)}")
```

## 3. Pascal Üçgeni

Her eleman, üstündeki iki elemanın toplamıdır ve binom katsayılarına karşılık gelir.

```python
def pascal_ucgeni(satir_sayisi):
    ucgen = []
    for n in range(satir_sayisi):
        ucgen.append([comb(n, k) for k in range(n + 1)])
    return ucgen

for satir in pascal_ucgeni(6):
    print(" ".join(f"{x:>3}" for x in satir).center(40))
```

## 4. Güvercin Yuvası İlkesi

Eğer $n$ nesne $m$ kutuya yerleştirilirse ve $n > m$ ise, en az bir kutuda birden fazla nesne bulunur.

```python
def guvercin_yuvasi_min_cakisma(nesneler, kutular):
    # En dolu kutuda garanti edilen minimum nesne sayısı
    return -(-nesneler // kutular)  # yukarı yuvarlama

print("13 kişi, 12 ay -> en az",
      guvercin_yuvasi_min_cakisma(13, 12), "kişi aynı doğum ayını paylaşır")
```
