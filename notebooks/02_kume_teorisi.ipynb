# Küme Teorisi

Kümeler, küme işlemleri, Venn şeması kavramları, kuvvet kümeleri ve Kartezyen çarpımlar.

## 1. Temel Küme İşlemleri

Birleşim, kesişim, fark ve simetrik fark.

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print("A ∪ B :", A | B)
print("A ∩ B :", A & B)
print("A - B :", A - B)
print("A △ B :", A ^ B)
```

## 2. Alt Kümeler ve Kuvvet Kümeleri

Bir $S$ kümesinin **kuvvet kümesi**, $S$'in tüm alt kümelerinin kümesidir. Eğer $|S| = n$ ise $|P(S)| = 2^n$ olur.

```python
from itertools import chain, combinations

def kuvvet_kumesi(s):
    s = list(s)
    return list(chain.from_iterable(combinations(s, r) for r in range(len(s) + 1)))

S = {1, 2, 3}
kk = kuvvet_kumesi(S)
print(f"{S} kümesinin kuvvet kümesi:")
for alt_kume in kk:
    print(set(alt_kume) if alt_kume else "∅")
print(f"\nBoyut: {len(kk)} = 2^{len(S)}")
```

## 3. Kartezyen Çarpım

Kartezyen çarpım $A \\times B$, $a \\in A$ ve $b \\in B$ olmak üzere tüm $(a, b)$ sıralı ikililerinin kümesidir.

```python
from itertools import product as kartezyen

A = {1, 2}
B = {'x', 'y', 'z'}
carpim = list(kartezyen(A, B))
print("A × B :", carpim)
print("|A × B| =", len(carpim), "= |A| · |B| =", len(A) * len(B))
```

## 4. Küme Özdeşlikleri

Evrensel küme $U$ içinde kümeler için De Morgan kurallarını doğrulayalım.

```python
U = set(range(1, 11))
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}

sol = U - (A | B)
sag = (U - A) & (U - B)
print("(A ∪ B)' =", sol)
print("A' ∩ B'  =", sag)
print("De Morgan geçerli:", sol == sag)
```
