# Çizge (Graf) Teorisi

Çizgeler, gösterimler, dolaşmalar (BFS/DFS), bağlılık ve temel özellikler.

## 1. Çizge Gösterimi

Bir çizgeyi komşuluk listesi (her düğümü komşularına eşleyen bir sözlük) kullanarak gösteririz.

```python
cizge = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E'],
}

def derece(cizge, dugum):
    return len(cizge[dugum])

for v in cizge:
    print(f"derece({v}) = {derece(cizge, v)}")

toplam_derece = sum(derece(cizge, v) for v in cizge)
print(f"\nDerecelerin toplamı = {toplam_derece} = 2 × |kenar sayısı|")
```

## 2. Genişlik Öncelikli Arama (BFS)

BFS, bir kuyruk kullanarak düğümleri seviye seviye dolaşır.

```python
from collections import deque

def bfs(cizge, baslangic):
    ziyaret = []
    gorulen = {baslangic}
    kuyruk = deque([baslangic])
    while kuyruk:
        dugum = kuyruk.popleft()
        ziyaret.append(dugum)
        for komsu in cizge[dugum]:
            if komsu not in gorulen:
                gorulen.add(komsu)
                kuyruk.append(komsu)
    return ziyaret

print("A'dan BFS:", bfs(cizge, 'A'))
```

## 3. Derinlik Öncelikli Arama (DFS)

DFS, geri dönmeden önce her dalı mümkün olduğunca derine kadar dolaşır.

```python
def dfs(cizge, baslangic, ziyaret=None):
    if ziyaret is None:
        ziyaret = []
    ziyaret.append(baslangic)
    for komsu in cizge[baslangic]:
        if komsu not in ziyaret:
            dfs(cizge, komsu, ziyaret)
    return ziyaret

print("A'dan DFS:", dfs(cizge, 'A'))
```

## 4. Bağlılık

Bir çizge, her düğüm çifti arasında bir yol varsa **bağlıdır**.

```python
def bagli_mi(cizge):
    baslangic = next(iter(cizge))
    erisilebilir = set(bfs(cizge, baslangic))
    return erisilebilir == set(cizge.keys())

print("Çizge bağlı mı:", bagli_mi(cizge))
```

## Alıştırmalar

1. Yönsüz bir çizgenin döngü içerip içermediğini tespit eden bir fonksiyon yazın.
2. Bir çizgedeki bağlı bileşen sayısını sayın.
3. BFS renklendirmesi kullanarak verilen bir çizgenin iki parçalı (bipartite) olup olmadığını belirleyin.
