# Mantık ve İspatlar

Bu defter önermeler mantığını, mantıksal bağlaçları, doğruluk tablolarını ve ayrık matematikteki ispat tekniklerini kapsar.

## 1. Önermeler Mantığı

Bir **önerme**, doğru ya da yanlış olan bildirim cümlesidir. Önermeleri mantıksal bağlaçlarla birleştiririz: DEĞİL (¬), VE (∧), VEYA (∨), İSE (→) ve ÇİFT YÖNLÜ KOŞUL (↔).

```python
from itertools import product

def dogruluk_tablosu(degiskenler, ifade):
    """Bir boolean ifade için doğruluk tablosu yazdırır.

    degiskenler: değişken adlarının listesi
    ifade: değişken->bool sözlüğü alıp bool döndüren fonksiyon
    """
    baslik = degiskenler + ["sonuc"]
    print(" | ".join(f"{h:>6}" for h in baslik))
    print("-" * (9 * len(baslik)))
    for degerler in product([False, True], repeat=len(degiskenler)):
        ortam = dict(zip(degiskenler, degerler))
        sonuc = ifade(ortam)
        satir = [str(ortam[v]) for v in degiskenler] + [str(sonuc)]
        print(" | ".join(f"{c:>6}" for c in satir))

# Örnek: koşullu önerme p -> q
dogruluk_tablosu(["p", "q"], lambda e: (not e["p"]) or e["q"])
```

## 2. Mantıksal Denklikler

İki önerme, her durumda aynı doğruluk değerine sahipse **mantıksal olarak denktir**. De Morgan kuralını doğrulayalım: ¬(p ∧ q) ≡ ¬p ∨ ¬q.

```python
def denk_mi(degiskenler, ifade1, ifade2):
    for degerler in product([False, True], repeat=len(degiskenler)):
        ortam = dict(zip(degiskenler, degerler))
        if ifade1(ortam) != ifade2(ortam):
            return False
    return True

de_morgan_sol = lambda e: not (e["p"] and e["q"])
de_morgan_sag = lambda e: (not e["p"]) or (not e["q"])

print("De Morgan kuralı geçerli:", denk_mi(["p", "q"], de_morgan_sol, de_morgan_sag))
```

## 3. Totolojiler ve Çelişkiler

Bir **totoloji** her zaman doğrudur; bir **çelişki** her zaman yanlıştır.

```python
def siniflandir(degiskenler, ifade):
    sonuclar = [ifade(dict(zip(degiskenler, v)))
                for v in product([False, True], repeat=len(degiskenler))]
    if all(sonuclar):
        return "totoloji"
    if not any(sonuclar):
        return "çelişki"
    return "olumsal (contingency)"

print("p ∨ ¬p :", siniflandir(["p"], lambda e: e["p"] or not e["p"]))
print("p ∧ ¬p :", siniflandir(["p"], lambda e: e["p"] and not e["p"]))
```

## 4. İspat Teknikleri

Yaygın teknikler: **doğrudan ispat**, **karşıt tersine göre ispat** ve **çelişki ile ispat**. Aşağıda biçimsel bir argümandan önce bir iddiayı hesaplamalı olarak doğruluyoruz.

**İddia:** Eğer $n$ çift ise $n^2$ de çifttir.

**Doğrudan ispat:** $n = 2k$ olsun. O hâlde $n^2 = 4k^2 = 2(2k^2)$ olur ki bu çifttir. ∎

```python
def n_kare_cift_kontrol(sinir=1000):
    for n in range(sinir):
        if n % 2 == 0 and (n * n) % 2 != 0:
            return False
    return True

print("Deneysel kontrol başarılı:", n_kare_cift_kontrol())
```
