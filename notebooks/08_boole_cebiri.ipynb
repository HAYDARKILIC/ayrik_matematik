# Boole Cebiri

Boole fonksiyonları, sadeleştirme, normal formlar ve mantık kapıları.

## 1. Boole Fonksiyonları ve Kapılar

Temel kapılar: VE (AND), VEYA (OR), DEĞİL (NOT), VEDEĞİL (NAND), VEYADEĞİL (NOR), ÖZELVEYA (XOR).

```python
def VE(a, b):    return a and b
def VEYA(a, b):  return a or b
def DEGIL(a):    return not a
def VEDEGIL(a, b): return not (a and b)
def OZELVEYA(a, b): return a != b

from itertools import product
print("a  b | VE VEYA XOR NAND")
for a, b in product([0, 1], repeat=2):
    print(f"{a}  {b} |  {int(VE(a,b))}   {int(VEYA(a,b))}    {int(OZELVEYA(a,b))}    {int(VEDEGIL(a,b))}")
```

## 2. Ayrık Normal Form (DNF)

Her Boole fonksiyonu, doğruluk tablosundan türetilen VE'lerin VEYA'sı (çarpımların toplamı) olarak ifade edilebilir.

```python
def dnf_ye_cevir(degiskenler, fonk):
    terimler = []
    for degerler in product([0, 1], repeat=len(degiskenler)):
        if fonk(*degerler):
            edebiyatlar = []
            for deg, val in zip(degiskenler, degerler):
                edebiyatlar.append(deg if val else f"¬{deg}")
            terimler.append("(" + " ∧ ".join(edebiyatlar) + ")")
    return " ∨ ".join(terimler) if terimler else "0"

# Örnek: iki değişkenin XOR'u
print("XOR DNF:", dnf_ye_cevir(["a", "b"], lambda a, b: a != b))
```

## 3. Fonksiyonel Tamlık

Tek başına NAND tüm Boole fonksiyonlarını ifade edebilir. NAND'den DEĞİL, VE ve VEYA inşa ediyoruz.

```python
def nand_den_degil(a):       return VEDEGIL(a, a)
def nand_den_ve(a, b):       return VEDEGIL(VEDEGIL(a, b), VEDEGIL(a, b))
def nand_den_veya(a, b):     return VEDEGIL(VEDEGIL(a, a), VEDEGIL(b, b))

for a, b in product([0, 1], repeat=2):
    assert nand_den_ve(a, b) == VE(a, b)
    assert nand_den_veya(a, b) == VEYA(a, b)
print("NAND fonksiyonel olarak tamdır: doğrulandı ✓")
```
